#Q1 ANSWER

create a User-defined Exception, we need to create a class directly or indirectly derived from the built-in Exception class.

Exception class is base class and child class are the derived class .Derived class inherit all the  method and functions from base class



```python
class percentagechecker(Exception):
    def __init__(self,msg):
        self.msg=msg
```

```python
def eligible(n):
    if n>100:
        raise percentagechecker("Invalid percentage")
    elif n<35:
        raise percentagechecker("You are failed so not eligible")
    elif n<60:
        raise percentagechecker("You are eligible but chances of selection is very very less")
    else:
        print("congratulation")
```

```python
try:
    n=int(input("enter your percentage"))
    eligible(n)
except percentagechecker as e:
    print(e)
```

ArithmeticError is simply an error that occurs during numeric calculations.

ArithmeticError types in Python include:

#Q4 ANSWER

The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively.

You can use LookupError exception class to handle both IndexError and KeyError exception classes.


```python
try:
    x=[10,20,30,40]
    x[10]
except LookupError as e:
    print(e,e.__class__)
```

```python
try:
    x=(1,2,3,4,5,6)
    x[9]
except LookupError as e:
    print(e,e.__class__)
```

```python
try:
    d={
        "name":"devvrat",
        "age":24,
        "status":"unmarried"
    }
    d["wife"]
except LookupError as e:
    print(e,e.__class__)
```

#Q5 ANSWER

In Python, when we want to include the module contents in the program then we have to import these specific modules in the program. So to do this we use “import” keyword such as import statement with the module name. When writing this statement and the specified module is not written properly or the imported module is not found in the Python library then the Python interpreter throws an error known as ImportError.



```python
try:
    import request
except ImportError as e:
    print(e,e.__class__)
```

In the above sample code, we can see that we are importing a module named “request” which is not present in the downloaded Python library. Therefore it throws an ImportError which gives the message saying no module named “request

Using try, except, and finally blocks

A “try” block encloses the code that may raise an exception, while one or more “except” blocks catch and handle specific exceptions. An optional “finally” block can be added to execute code regardless of whether an exception was raised or not.



```python
def divide(a,b):
    try:
        result=a/b
    except ZeroDivisionError as e:
        print("OOps there is an erro")
        print(e)
    else:
        print(f"the result is {result}")
    finally:
        print("the divide function has been executed")
```

```python
divide(4,2)
```

```python
divide(4,0)
```

Catching specific exceptions

Catching specific exceptions is crucial for writing maintainable and easy-to-debug code. By specifying the exact exception types you expect to handle, you can avoid inadvertently suppressing unexpected exceptions that may reveal underlying issues in your code.


```python
try:
    f=open("alakh.txt","r")
    f.read()
except Exception as e:
    print(e,e.__class__)
```
